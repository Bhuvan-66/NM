```python
# Step 1: Import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Step 2: Create dataset
data = {
    "Product": ["Laptop", "Mobile", "Laptop", "Tablet", "Mobile", "Tablet", "Laptop"],
    "Quantity": [5, 10, 3, 7, 8, 2, 6],
    "Price": [50000, 20000, None, 15000, None, 12000, 52000]
}

df = pd.DataFrame(data)

# Step 3: Fill missing Price values with product-wise average
df["Price"] = df.groupby("Product")["Price"].transform(lambda x: x.fillna(x.mean()))

# Step 4: Add Total_Sales column
df["Total_Sales"] = df["Quantity"] * df["Price"]

# Step 5: Group by Product and calculate total sales
sales_summary = df.groupby("Product")["Total_Sales"].sum()

# Step 6: Find product with highest sales
top_product = sales_summary.idxmax()
print("Product with highest total sales:", top_product)

# Step 7: Visualization
plt.figure()
sales_summary.plot(kind='bar')
plt.title("Total Sales by Product")
plt.xlabel("Product")
plt.ylabel("Total Sales")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
df = pd.DataFrame({
    "Date": ["2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04", "2024-01-05"],
    "Min_Temp": [20, 22, None, 21, 23],
    "Max_Temp": [30, None, 35, 33, 36]
})

# FIXED: Fill missing values (no inplace)
df["Min_Temp"] = df["Min_Temp"].fillna(df["Min_Temp"].mean())
df["Max_Temp"] = df["Max_Temp"].fillna(df["Max_Temp"].mean())

# Average temperature
df["Average_Temp"] = (df["Min_Temp"] + df["Max_Temp"]) / 2

# Highest temperature date
max_date = df.loc[df["Average_Temp"].idxmax(), "Date"]
print("Highest Average Temperature Date:", max_date)

# Plot
plt.plot(df["Date"], df["Average_Temp"], marker='o')
plt.title("Average Temperature Trend")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
df = pd.DataFrame({
    "Date": ["2024-01-01", "2024-01-02", "2024-01-03", "2024-01-04", "2024-01-05"],
    "Vehicle_Count": [120, 150, None, 180, 200]
})

# Replace missing values with 0
df["Vehicle_Count"] = df["Vehicle_Count"].fillna(0)

# Calculate total and average
total = df["Vehicle_Count"].sum()
average = df["Vehicle_Count"].mean()

print("Total Vehicle Count:", total)
print("Average Vehicle Count:", average)

# Find date with highest traffic
max_date = df.loc[df["Vehicle_Count"].idxmax(), "Date"]
print("Highest Traffic Date:", max_date)

# Plot line graph
plt.plot(df["Date"], df["Vehicle_Count"], marker='o')
plt.title("Daily Traffic Volume")
plt.xlabel("Date")
plt.ylabel("Vehicle Count")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
df = pd.DataFrame({
    "Student_ID": [101, 102, 103, 104, 105, 106, 107],
    "Completion_Status": ["Yes", "No", None, "Yes", "No", None, "Yes"]
})

# Fill missing values with "No"
df["Completion_Status"] = df["Completion_Status"].fillna("No")

# Count Yes and No
counts = df["Completion_Status"].value_counts()
print(counts)

# Pie chart
plt.pie(counts, labels=counts.index, autopct='%1.1f%%')
plt.title("Course Completion Status")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create dataset
df = pd.DataFrame({
    "Product_Name": ["Laptop", "Mobile", "Tablet", "Headphones", "Camera", "Watch", "Speaker"],
    "Customer_Rating": [4.5, 4.2, None, 4.8, 4.6, 4.1, 4.7]
})

# Remove rows with missing ratings
df = df.dropna()

# Calculate average rating
avg_rating = df["Customer_Rating"].mean()
print("Average Rating:", avg_rating)

# Top 3 highest-rated products
top3 = df.sort_values(by="Customer_Rating", ascending=False).head(3)
print("\nTop 3 Products:\n", top3)

# Top 5 products for visualization
top5 = df.sort_values(by="Customer_Rating", ascending=False).head(5)

# Bar chart
plt.bar(top5["Product_Name"], top5["Customer_Rating"])
plt.title("Top 5 Product Ratings")
plt.xlabel("Product")
plt.ylabel("Rating")
plt.show()
```
